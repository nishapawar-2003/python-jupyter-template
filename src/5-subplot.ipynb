### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.linspace(0, 10, 100)
y_line = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [25, 15, 30, 20]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 1, 5, 3]
labels_pie = ['Category 1', 'Category 2', 'Category 3']
sizes_pie = [40, 35, 25]

# Create the subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns

# Plot 1: Line chart
axes[0, 0].plot(x, y_line)
axes[0, 0].set_title('Line Chart')

# Plot 2: Bar chart
axes[0, 1].bar(categories, values)
axes[0, 1].set_title('Bar Chart')

# Plot 3: Scatter plot
axes[1, 0].scatter(x_scatter, y_scatter)
axes[1, 0].set_title('Scatter Plot')

# Plot 4: Pie chart
axes[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90)
axes[1, 1].axis('equal')  # Equal aspect ratio for a circle
axes[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()

# Show the plot
plt.show()
```
