### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(6, 6))  # Adjust figure size for better annotation visibility if needed

plt.scatter(x, y)

for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')  # Annotate each point

plt.title('Scatter Plot with Annotations')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True) # Add a grid for better readability (optional)
plt.show()
```
